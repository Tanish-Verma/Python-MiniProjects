```python
import sqlite3
```

```python
conn = sqlite3.connect("customer.db")

c = conn.cursor()
```

```python
#create a table

c.execute("""
    CREATE TABLE customers(
        first_name text,
        last_name text,
        Email text
)""")

conn.commit()

#close the connection
```

```python
c.execute("""
    INSERT INTO customers VALUES (
               'Mary',
               'hail',
               'mh@codemy.com'
               )""")

print("Command executed succesfully!...")
conn.commit()
```

```python
c.execute("""
    INSERT INTO customers VALUES (
               'Mary',
               'Brown',
               'mary@codemy.com'
               )""")

print("Command executed succesfully!...")
conn.commit()
```

```python
c.execute("SELECT * FROM customers")
# print(c.fetchone()[1])
# c.fetchmany(3)
# print(c.fetchall())

items = c.fetchall()

print(items)
conn.commit()
```

```python
c.execute("SELECT rowid, * FROM customers")

items = c.fetchall()
for item in items:
    print(item)
```

```python
c.execute("SELECT rowid, * FROM customers WHERE email LIKE '%@codemy%'")

items = c.fetchall()
for item in items:
    print(item)
```

```python
c.execute("""
    UPDATE customers SET first_name ='bob'
    WHERE last_name = 'Elder'
""")
c.execute("SELECT * FROM customers")

items = c.fetchall()
for item in items:
    print(item)
conn.commit()
```

```python
c.execute("""
    UPDATE customers SET first_name ='john'
    WHERE rowid = 1
""")
c.execute("SELECT rowid,* FROM customers")

items = c.fetchall()
for item in items:
    print(item)
conn.commit()
```

```python
c.execute("DELETE from customers WHERE rowid = 7")
c.execute("SELECT rowid,* FROM customers")
items = c.fetchall()
for item in items:
    print(item)
conn.commit()
```

```python
#order by
c.execute("SELECT rowid,* FROM customers ORDER BY last_name")
items = c.fetchall()
for item in items:
    print(item)
conn.commit()
```

```python
#order by
c.execute("SELECT rowid,* FROM customers LIMIT 2")
items = c.fetchall()
for item in items:
    print(item)
conn.commit()
```

```python
c.execute("DROP TABLE customers")
conn.commit()
```
